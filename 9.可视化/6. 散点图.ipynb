```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df[:5]
```

**散点图需要x和y两个参数**

```python
df.plot.scatter(x='a', y='b')
```

**要在一个轴上绘制多个组, 只需要在同一个ax上多次plot**

```python
ax = df.plot.scatter(x='a', y='c', color='red', label='a-c')
df.plot.scatter(x='c', y='b', color='blue', label='b-c', ax=ax)
```

**参数c(列表形式)设置每个点的颜色**

```python
df.plot.scatter(x='a', y='b', c='c')
```

**参数s(列表形式)设置每个点的半径大小**

```python
df.plot.scatter(x='a', y='b', s=df['c']*50)
```
