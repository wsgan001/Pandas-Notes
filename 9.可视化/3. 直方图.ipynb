```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
df = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000),  'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
df[:5]
```

**DataFrame上绘制直方图**

```python
df.plot(kind='hist', alpha=.7)  # alpha: 透明度
```

**Series上绘制直方图**

```python
df['a'].plot(kind='hist')
```

**堆叠直方图设置stacked=True**

```python
df.plot(kind='hist', stacked=True)
```

**通过设置bins参数修改子区间数量**

```python
df.plot(kind='hist', stacked=True, bins=30)
```

```python
df.plot(kind='hist', stacked=True, bins=20)
```

```python
df.plot(kind='hist', stacked=True, bins=10)
```

**还可以传递matplotlib中hist支持的其他任意参数**

如
- 水平: orientation='horizontal'
- 累积: cumulative='True'

```python
df.plot(kind='hist', orientation='horizontal')
```

**DataFrame.hist()在多个子图上绘制列的直方图**

```python
df.hist()
```

**by参数绘制每个分组**

在DataFrame上分组

```python
df.hist(by=np.random.randint(0, 3, len(df)))
```

在Series上分组

```python
df['a'].hist(by=np.random.randint(0, 3, len(df)))
```
