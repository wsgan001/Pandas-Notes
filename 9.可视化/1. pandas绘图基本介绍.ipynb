1. **plot方法只是plt.plot()的一个简单包装**

**绘制Series数据**

```python
# 如果索引有日期构成, 则会调用gcf().autofmt_xdate()来格式化x轴

ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

**绘制DataFrame数据**

在DataFrame上, plot()绘制所有带标签的列

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
df.plot()
```

可以使用plot()中的参数x和y来绘制DataFrame中的两列, 一列作为x, 一列作为y

```python
df.plot(x='B', y='C')
```

2.**绘图方法除了允许默认线图外, 还有一些别的可以通过plot()里的kind参数来控制的图**

**也可以通过plot.kind()来创建这些图, 二者等价**


- bar或barh
- hist
- box
- kde或density
- area
- scatter
- hexbin
- pie

```python
pd.Series(np.random.randn(10)).plot(kind='bar')
```

```python
pd.Series(np.random.randn(10)).plot.bar()
```
